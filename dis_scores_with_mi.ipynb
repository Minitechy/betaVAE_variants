### 1. MIG and I10 Score

```python
# Helper functions
def compute_joint_probabilities(x, y):
    """
    Compute the joint probability matrix for two discrete variables.
    """
    joint_hist = np.histogram2d(x, y, bins=(np.unique(x).size, np.unique(y).size))[0]
    joint_prob = joint_hist / np.sum(joint_hist)  # Normalize to get probabilities
    return joint_prob

def compute_marginal_probabilities(joint_prob):
    """
    Compute the marginal probabilities P(x) and P(y) from the joint probability matrix.
    """
    marginal_x = np.sum(joint_prob, axis=1)  # P(x)
    marginal_y = np.sum(joint_prob, axis=0)  # P(y)
    return marginal_x, marginal_y

def get_mutual_information(x, y, eps=1e-10):
    """
    Compute mutual information between two discrete variables.
    """
    joint_prob = compute_joint_probabilities(x, y)
    marginal_x, marginal_y = compute_marginal_probabilities(joint_prob)
    
    mi = 0.0
    for i in range(joint_prob.shape[0]):
        for j in range(joint_prob.shape[1]):
            if joint_prob[i, j] > 0:  # Avoid log(0) issues
                mi += joint_prob[i, j] * np.log((joint_prob[i, j] + eps) / (marginal_x[i] * marginal_y[j] + eps))
    return mi

def compute_entropy(x):
    """
    Compute the entropy of a discrete variable.
    """
    value_counts = np.bincount(x)
    probabilities = value_counts / len(x)
    entropy = -np.sum(probabilities[probabilities > 0] * np.log(probabilities[probabilities > 0]))
    return entropy

# Quantile-based discretization of continuous latent variables
def quantile_based_discretize(latents, nb_bins):
    """
    Discretize continuous latent variables using quantile-based binning (equal frequency).
    """
    discretizer = KBinsDiscretizer(n_bins=nb_bins, encode='ordinal', strategy='quantile')
    latents_discretized = discretizer.fit_transform(latents)
    return latents_discretized.astype(int)

# MIG computation with added plot generation and latent space dimension selection
def mig_with_plot(factors, latents, nb_bins, beta, lambda_reg, mi_flag):
    """
    Compute the MIG score, display a heatmap of MI between factors and latents, and save the plot.
    Additionally, print which latent space dimension captures each ground truth factor.
    
    Parameters:
    - factors: Discrete ground truth factors (shape, scale, orientation, posX, posY).
    - latents: Continuous latent variables.
    - nb_bins: Number of bins to discretize latent variables.
    - beta: The fixed beta value for the computation.
    - lambda_reg: The lambda value being processed.
    """
    nb_factors = factors.shape[1]
    nb_latents = latents.shape[1]

    # Discretize continuous latent variables using quantile-based binning
    latents_discretized = quantile_based_discretize(latents, nb_bins)

    # Compute mutual information (MI) matrix
    mi_matrix = np.zeros((nb_factors, nb_latents))
    for f in range(nb_factors):
        for l in range(nb_latents):
            mi_matrix[f, l] = get_mutual_information(factors[:, f], latents_discretized[:, l])

    # Compute entropy of each factor
    entropy_factors = np.array([compute_entropy(factors[:, f]) for f in range(nb_factors)])  
    
    # Plot the MI matrix as a heatmap
    plt.figure(figsize=(10, 6))
    ax = sns.heatmap(mi_matrix, annot=True, fmt=".2f", cmap='Blues', cbar_kws={'label': 'Mutual Information'})

    ax.set_xlabel('Latent Space Dimensions', fontsize=12)
    ax.set_ylabel('Ground Truth Factors', fontsize=12)
    ax.set_xticks(np.arange(nb_latents) + 0.5)
    ax.set_yticks(np.arange(nb_factors) + 0.5)

    ax.set_xticklabels([f'$x_{{{i+1}}}$' for i in range(nb_latents)], fontsize=10)
    ax.set_yticklabels(['Shape', 'Scale', 'Orientation', 'PosX', 'PosY'], fontsize=10)
    
    # Save the plot
    plt.tight_layout()
    
    if mi_flag == 1:
        plt.savefig(f"{SAVE_PATH}val_mutual_info_with_mi_beta_{beta}_lambda_{lambda_reg}.png", bbox_inches='tight')
    else:
        plt.savefig(f"{SAVE_PATH}val_mutual_info_without_mi_beta_{beta}_lambda_{lambda_reg}.png", bbox_inches='tight')
        
    plt.show()

    # Print which latent dimension captures each ground truth factor the best
    if mi_flag == 1:
        print(colored(f"\nLatent Dimension Mapping with mi for beta = {beta}, lambda = {lambda_reg}:", attrs=['bold']))
    else:
        print(colored(f"\nLatent Dimension Mapping without mi for beta = {beta}, lambda = {lambda_reg}:", attrs=['bold']))

    latent_mapping = []
    mig_per_factor = []
    for f in range(nb_factors):
        best_latent_idx = np.argmax(mi_matrix[f, :])
        best_mi = mi_matrix[f, best_latent_idx]
        latent_mapping.append(best_latent_idx)
        mig_per_factor.append(best_mi)
        print(f"'{['Shape', 'Scale', 'Orientation', 'PosX', 'PosY'][f]}' is best captured by x_{best_latent_idx + 1}")

    # Compute the MIG score
    sum_gap = 0
    I10_score = 0
    
    for f in range(nb_factors):
        mi_f_sorted = np.sort(mi_matrix[f, :])[::-1]  # Sort MI for factor f in descending order
        gap = mi_f_sorted[0] - mi_f_sorted[1]  # Gap between highest and second-highest MI
        sum_gap += gap / entropy_factors[f]  # Normalize by the entropy of the factor  

        # I10 score
        I10_score += mi_f_sorted[0]
        
    mig_score = round(sum_gap / nb_factors, 4)  # Average across all factors
    
    return mig_score, latent_mapping, mig_per_factor, round(I10_score, 4)

# MIG and I10 computation across lambdas with fixed beta
def compute_mig_across_lambdas(factors_list, latents_list, nb_bins, beta, lambdas, mi_flag):
    """
    Compute MIG scores across different lambda values for a fixed beta, save plots, and print latent dimension mapping.

    Parameters:
    - factors_list: List of ground truth factors for each lambda.
    - latents_list: List of latent variables for each lambda.
    - nb_bins: Number of bins for discretization.
    - beta: Fixed beta value.
    - lambdas: List of lambda values.
    """
    mig_scores = []
    I10_scores = []
    latent_mappings = []  # Store the best latent dimension for each factor across lambdas
    mig_matrix = []  # Store MIG scores for each factor across lambdas

    # Compute MIG and I10 for each lambda and visualize the MI matrix
    for i, lambda_reg in enumerate(lambdas):
        if mi_flag == 1:
            print(colored(f"\nComputing MIG and I10 with mi for beta = {beta}, lambda = {lambda_reg}...", "red", attrs=['bold']))
        else:
            print(colored(f"\nComputing MIG and I10 without mi for beta = {beta}, lambda = {lambda_reg}...", "red", attrs=['bold']))
        
        mig_score, latent_mapping, mig_per_factor, I10_score = mig_with_plot(factors_list[i], latents_list[i], nb_bins, beta, lambda_reg, mi_flag)

        mig_scores.append(mig_score)
        latent_mappings.append(latent_mapping)
        mig_matrix.append(mig_per_factor)

        if mi_flag == 1:
            print(colored(f"\nMIG Score: {mig_score}", "blue", attrs=['bold']))
        else:
            print(colored(f"\nMIG Score: {mig_score}", "green", attrs=['bold']))

        I10_scores.append(I10_score)

        if mi_flag == 1:
            print(colored(f"I10 Score: {I10_score}", "blue", attrs=['bold']))
        else:
            print(colored(f"I10 Score: {I10_score}", "green", attrs=['bold']))

    # Create a matrix for the heatmap that includes the best latent dimension and MIG score
    factor_names = ['Shape', 'Scale', 'Orientation', 'PosX', 'PosY']
    nb_factors = len(factor_names)
    heatmap_data = np.empty((nb_factors, len(lambdas)), dtype=object)
    mi_values = np.zeros((nb_factors, len(lambdas)))  # To store MI scores for the color bar

    for f in range(nb_factors):
        for l in range(len(lambdas)):
            best_latent_idx = latent_mappings[l][f]  # Best latent dimension for this factor
            mig_value = mig_matrix[l][f]  # MIG score for this factor
            heatmap_data[f, l] = f'$x_{{{best_latent_idx + 1}}}$\n({mig_value:.2f})'  # Latent dimension + MIG score
            mi_values[f, l] = mig_value  # Store MI score for the color bar

    # Plot the heatmap for best latent dimensions capturing each factor
    plt.figure(figsize=(10, 6))

    # Using mi_values for the color mapping and heatmap background, annotating with the best latent dimension
    ax = sns.heatmap(mi_values, annot=heatmap_data, fmt='', cmap='Reds', cbar_kws={'label': 'Mutual Information'})

    ax.set_xlabel(r'$\lambda$', fontsize=12)
    ax.set_ylabel('Ground Truth Factors', fontsize=12)
    ax.set_xticks(np.arange(len(lambdas)) + 0.5)
    ax.set_yticks(np.arange(nb_factors) + 0.5)

    ax.set_xticklabels([f'{lambda_reg}' for lambda_reg in lambdas], fontsize=10)
    ax.set_yticklabels(factor_names, fontsize=10)

    # Save the heatmap
    plt.tight_layout()

    if mi_flag == 1:
        plt.savefig(f"{SAVE_PATH}val_latent_mapping_with_mi_fixed_beta_{beta}_across_lambdas.png", bbox_inches='tight')
    else:
        plt.savefig(f"{SAVE_PATH}val_latent_mapping_without_mi_fixed_beta_{beta}_across_lambdas.png", bbox_inches='tight')
        
    plt.show()

    return mig_scores, I10_scores
```

### 2. SAP Score

```python
# SAP score using covariance and variance, computed for each iteration
def sap(factors, latents, mi_flag):
    """
    Compute the SAP score using covariance and variance.
    
    Parameters:
    - factors: Continuous ground truth factors.
    - latents: Latent representations (encoder output).

    Returns:
    - sap_score: The computed SAP score based on the mean gap between the two most predictive latents for each factor.
    """
    nb_factors = factors.shape[1]
    nb_latents = latents.shape[1]

    # Initialize the R² score matrix (S) with dimensions [nb_latents, nb_factors]
    s_matrix = np.zeros((nb_latents, nb_factors))

    # Compute R² scores for each latent-factors pair using covariance and variance
    for f in range(nb_factors):
        for l in range(nb_latents):
            # Compute covariance between latent l and factor f
            cov = np.cov(latents[:, l], factors[:, f])[0, 1]
            
            # Compute variance of latent l and factor f
            var_latent = np.var(latents[:, l])
            var_factor = np.var(factors[:, f])
            
            # Compute R² score as the squared correlation
            if var_latent > 0 and var_factor > 0:  # Avoid division by zero
                r2_score = (cov / np.sqrt(var_latent * var_factor)) ** 2
            else:
                r2_score = 0
            
            # Store the R² score in the matrix
            s_matrix[l, f] = max(0, r2_score)

    # Compute the SAP score
    sap_total = 0
    for f in range(nb_factors):
        # Sort the R² scores for the f-th factor to find the top 2 most predictive latents
        sorted_r2_scores = np.sort(s_matrix[:, f])  # Sort in ascending order

        # The highest score and the second-highest score
        top_r2 = sorted_r2_scores[-1]  # Highest R² score
        second_top_r2 = sorted_r2_scores[-2]  # Second-highest R² score

        # Difference between the highest and second-highest R² score
        gap = top_r2 - second_top_r2
        sap_total += gap  # Accumulate the difference

    # Compute the mean gap (SAP score)
    sap_score = round(sap_total / nb_factors, 4)

    return sap_score

# SAP computation across different lambdas and betas
def compute_sap_across_lambdas_and_betas(factors_dict, latents_dict, betas, lambdas, mi_flag):
    """
    Compute SAP scores across multiple lambda and beta values.

    Parameters:
    - factors_dict: Dictionary of ground truth factors for each (beta, lambda) pair.
    - latents_dict: Dictionary of latent variables for each (beta, lambda) pair.
    - betas: List of beta values.
    - lambdas: List of lambda values.

    Returns:
    - sap_scores: Dictionary containing SAP scores for each (beta, lambda) pair.
    """
    sap_scores = {}

    # Compute SAP for each beta and lambda value
    for beta in betas:
        for lambda_reg in lambdas:

            if mi_flag == 1:
                print(colored(f"\nComputing SAP with mi for beta = {beta}, lambda = {lambda_reg}...", "blue", attrs=['bold']))
            else:
                print(colored(f"\nComputing SAP without mi for beta = {beta}, lambda = {lambda_reg}...", "green", attrs=['bold']))
                
            factors = factors_dict[(beta, lambda_reg)]
            latents = latents_dict[(beta, lambda_reg)]
            val_sap_score = sap(factors, latents, mi_flag)
            sap_scores[(beta, lambda_reg)] = val_sap_score
            print(colored(f"SAP Score: {val_sap_score}", attrs=['bold']))

    return sap_scores

# Dictionary to store the factors and latents data for validation set with different beta and lambda combinations
val_factors_dict_with_mi = {
    # Entries for beta = 6
    (6, 0.1): val_latent_values_with_mi_lambda_01_beta_6,
    (6, 0.2): val_latent_values_with_mi_lambda_02_beta_6,
    (6, 0.3): val_latent_values_with_mi_lambda_03_beta_6,
    (6, 0.4): val_latent_values_with_mi_lambda_04_beta_6,
    (6, 0.5): val_latent_values_with_mi_lambda_05_beta_6,
    # Entries for beta = 8
    (8, 0.1): val_latent_values_with_mi_lambda_01_beta_8,
    (8, 0.2): val_latent_values_with_mi_lambda_02_beta_8,
    (8, 0.3): val_latent_values_with_mi_lambda_03_beta_8,
    (8, 0.4): val_latent_values_with_mi_lambda_04_beta_8,
    (8, 0.5): val_latent_values_with_mi_lambda_05_beta_8
}

val_factors_dict_without_mi = {
    # Entries for beta = 6
    (6, 0.1): val_latent_values_without_mi_lambda_01_beta_6,
    (6, 0.2): val_latent_values_without_mi_lambda_02_beta_6,
    (6, 0.3): val_latent_values_without_mi_lambda_03_beta_6,
    (6, 0.4): val_latent_values_without_mi_lambda_04_beta_6,
    (6, 0.5): val_latent_values_without_mi_lambda_05_beta_6,
    # Entries for beta = 8
    (8, 0.1): val_latent_values_without_mi_lambda_01_beta_8,
    (8, 0.2): val_latent_values_without_mi_lambda_02_beta_8,
    (8, 0.3): val_latent_values_without_mi_lambda_03_beta_8,
    (8, 0.4): val_latent_values_without_mi_lambda_04_beta_8,
    (8, 0.5): val_latent_values_without_mi_lambda_05_beta_8
}

val_latents_dict_with_mi = {
    # Entries for beta = 6
    (6, 0.1): val_latent_vectors_with_mi_lambda_01_beta_6,
    (6, 0.2): val_latent_vectors_with_mi_lambda_02_beta_6,
    (6, 0.3): val_latent_vectors_with_mi_lambda_03_beta_6,
    (6, 0.4): val_latent_vectors_with_mi_lambda_04_beta_6,
    (6, 0.5): val_latent_vectors_with_mi_lambda_05_beta_6,
    # Entries for beta = 8
    (8, 0.1): val_latent_vectors_with_mi_lambda_01_beta_8,
    (8, 0.2): val_latent_vectors_with_mi_lambda_02_beta_8,
    (8, 0.3): val_latent_vectors_with_mi_lambda_03_beta_8,
    (8, 0.4): val_latent_vectors_with_mi_lambda_04_beta_8,
    (8, 0.5): val_latent_vectors_with_mi_lambda_05_beta_8
}

val_latents_dict_without_mi = {
    # Entries for beta = 6
    (6, 0.1): val_latent_vectors_without_mi_lambda_01_beta_6,
    (6, 0.2): val_latent_vectors_without_mi_lambda_02_beta_6,
    (6, 0.3): val_latent_vectors_without_mi_lambda_03_beta_6,
    (6, 0.4): val_latent_vectors_without_mi_lambda_04_beta_6,
    (6, 0.5): val_latent_vectors_without_mi_lambda_05_beta_6,
    # Entries for beta = 8
    (8, 0.1): val_latent_vectors_without_mi_lambda_01_beta_8,
    (8, 0.2): val_latent_vectors_without_mi_lambda_02_beta_8,
    (8, 0.3): val_latent_vectors_without_mi_lambda_03_beta_8,
    (8, 0.4): val_latent_vectors_without_mi_lambda_04_beta_8,
    (8, 0.5): val_latent_vectors_without_mi_lambda_05_beta_8
}

betas = [6, 8]

# Compute SAP scores for validation set across lambdas and betas
mi_flag = 1
sap_scores_dict_with_mi = compute_sap_across_lambdas_and_betas(val_factors_dict_with_mi, val_latents_dict_with_mi, betas, lambdas, mi_flag)

mi_flag = 0
sap_scores_dict_without_mi = compute_sap_across_lambdas_and_betas(val_factors_dict_without_mi, val_latents_dict_without_mi, betas, lambdas, mi_flag)
```
